```python

mark = float(input("Enter your marks: "))

if mark >= 80.00:
    print("Congratulations! A+")
elif 75.00 <= mark < 80.00:
    print("Congratulations! A")
elif 71.00 <= mark < 74.00:
    print("Congratulations! A-")
elif 65.00 <= mark < 70.00:
    print("Congratulations! B+")
elif 61.00 <= mark < 64.00:
    print("Congratulations! B")
elif 55.00 <= mark < 60.00:
    print("Congratulations! B-")
elif 51.00 <= mark < 54.00:
    print("Congratulations! C+")
elif 45.00 <= mark < 50.00:
    print("Congratulations! C")
elif 41.00 <= mark < 44.00:
    print("Congratulations! D")
else:
    print("Failed!")
```
