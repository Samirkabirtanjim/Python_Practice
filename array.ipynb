```python
from collections import deque

class BFS:
    def __init__(self):
        self.directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]
        self.found = False  # Initialize found attribute

    def st_bfs(self, graph, n, m):
        queue = deque([(self.source.x, self.source.y, 0)])
        visited = [[False for _ in range(m)] for _ in range(n)]
        visited[self.source.x][self.source.y] = True

        while queue:
            x, y, level = queue.popleft()

            if x == self.goal.x and y == self.goal.y:
                self.found = True
                self.goal_level = level
                break

            for dx, dy in self.directions:
                nx, ny = x + dx, y + dy

                if 0 <= nx < n and 0 <= ny < m and not visited[nx][ny] and graph[nx][ny] == 1:
                    queue.append((nx, ny, level + 1))
                    visited[nx][ny] = True

    def solve(self):
        n = int(input("Enter the number of rows in the grid: "))
        m = int(input("Enter the number of columns in the grid: "))

        class Node:
            def __init__(self, x, y, level, parent=None):
                self.x = x
                self.y = y
                self.level = level
                self.parent = parent

        graph = []
        for i in range(n):
            row = [int(x) for x in input(f"Enter the elements of row {i + 1}: ").split()]
            graph.append(row)

        source_x = int(input("Enter the x-coordinate of the start state: "))
        source_y = int(input("Enter the y-coordinate of the start state: "))
        goal_x = int(input("Enter the x-coordinate of the goal state: "))
        goal_y = int(input("Enter the y-coordinate of the goal state: "))

        self.source = Node(source_x, source_y, 0)
        self.goal = Node(goal_x, goal_y, float('inf'))

        self.st_bfs(graph, n, m)

        if self.found:
            print("Goal found")
            print("Number of moves required =", self.goal_level)
            path = []
            x, y = self.goal.x, self.goal.y

            while x != self.source.x or y != self.source.y:
                path.append((x, y))
                if self.goal.parent is not None:
                    nx, ny = x - self.goal.parent[0], y - self.goal.parent[1]
                    x, y = nx, ny
                else:
                    break

            path.append((x, y))

            print("Path:")
            for i in range(len(path) - 1, -1, -1):
                print(f"({path[i][0]}, {path[i][1]})")

        else:
            print("Goal cannot be reached from starting block")


if __name__ == "__main__":
    bfs = BFS()
    bfs.solve()
```
