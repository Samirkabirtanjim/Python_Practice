# **half pyramid a using numbers**

```python
num = int(input("Enter a number: "))
for row in range(0,num+1):
  for col in range(1,row+1):
    print(col, end="")
  print()
```

# **Inverted half pyramid using numbers**

```python
num = int(input("Enter a number: "))
for row in range(num, 0, -1):
  for col in range(1,row+1):
    print(col, end="")
  print()
```
