```python
num = int(input("Enter the number of rows: "))
for row in range(0,num):
  for column in range(0,row + 1):
    print(column,end="")
  print()
```

```python
num = int(input("Enter the number of rows: "))
for row in range(0, num):
  for col in range(0, row+1):
    print("* ",end="")
  print()
```

```python
num = int(input("Enter the number of rows: "))
letter = ord('A')
for row in range(0, num):
  for col in range(0, row+1):
      a = chr(letter+col)
      print(a,end="")
  print()
```
