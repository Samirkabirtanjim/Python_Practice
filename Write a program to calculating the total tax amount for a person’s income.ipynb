```python
print("Welcome to Tax-Calculation")
print("\nTax Brackets:\n Income up to $10,000 is taxed at a rate of 5%.\n Income from $10,001 to $50,000 is taxed at a rate of 10%.\n Income from $50,001 to $100,000 is taxed at a rate of 20%.\n Income over $100,000 is taxed at a rate of 30%.")

income = float(input("Entaer your income: "))

if income <= 10000:
    tax = income * 0.05
    print("Your Tax:", tax)
elif income <= 50000:
    tax = 10000 * 0.05 + (income - 10000) * 0.10
    print("Your Tax:", tax)
elif income <= 100000:
    tax = 10000 * 0.05 + 40000 * 0.10 + (income - 50000) * 0.20
    print("Your Tax:", tax)
else:
    tax = 10000 * 0.05 + 40000 * 0.10 + 50000 * 0.20 + (income - 100000) * 0.30
    print("Your Tax:", tax)
```
