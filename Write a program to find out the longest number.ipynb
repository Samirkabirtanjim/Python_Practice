```python
x = input("Enter x number: ")
y = input("Enter y number: ")
z = input("Enter z number: ")

xx = len(str(x))
yy = len(str(y))
zz = len(str(z))

max_length = max(xx, yy, zz)

if xx == max_length:
  print(f"The longest number ",x)
elif yy == max_length:
  print(f"The longest number ",y)
else:
  print(f"The longest number ",z)
```
