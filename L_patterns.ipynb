# **Left To Right Pattern**

```python
num = int(input("Enter the number: "))
for row in range(1,num+1):
  for column in range(1,row+1):
    print("# ",end="")
  print()
```

# **Up To Down All Star**

```python
num = int(input("Enter the number: "))
for row in range(1,num+1):
  for column in range(1,num+1):
    print('*',end="")
  print()
```

# Large To **Small**

```python
num = int(input("Enter the number: "))
for row in range(0,num+1):
  for col in range(num-row,0,-1):
    print("#",end="")
  print()
```

# **1 To Mum **

```python
num = int(input("Enter the number: "))
for row in range(1,num+1):
  for col in range(1,row+1):
    print(col,end="")
  print()
```
